A2. Take 2 feature vectors which are independent but not orthogonal. Create 2 orthogonal vectors 
from these feature vectors using the following steps:
• Consider these vectors as v1 & v2
• Consider v1 as the initial vector (one of the orthogonal vectors)
• Find the projection vector of v2 onto v1. Call this p2. See below diagram for p2 calculation.
• Subtract the projection vector p2 from v2. Call the result of subtraction as h2.
• Check orthogonality between v1 & & h2.


```python
import math
v1=featurev1
v2=featurev2

# p2len = math.sqrt(np.dot(v2, v2)) *(np.dot(v1, v2) /(math.sqrt(np.dot(v1, v1))*math.sqrt(np.dot(v2, v2))) ) 

p2 = v1 * np.dot(v2, v1) / np.dot(v1, v1)

h2 = v2 - p2


if np.dot(v1,h2) != 0:
  print("yes they are orthogonal")
else:
  print("No they are not orthogonal")
```
