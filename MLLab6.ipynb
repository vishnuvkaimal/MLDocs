```python

def perceptron(x, w, b):
    yin = np.dot(w, x) + b
    y=1/(1+math.exp(-yin))
    # print(yin)
    return y
```

A1. Develop the above Neural Network. Use learning rate (Î±) = 0.05 with a Sigmoid activation 
function. Learn the weights of the network using back-propagation algorithm to implement above 
provided AND gate logic.

```python
import numpy as np
import matplotlib.pyplot as plt
import math


learnRate=0.05


v=[[0.3,0.5],[0.6,-0.3],[-0.1,0.4]]
w=[[-0.2],[0.4],[0.1]]


x=[[0,0],[0,1],[1,0],[1,1]]
t=[0,0,0,1]

# errEpoch=[]

count=0

for i in range(0,1000):
   e=0
   for j in range(0,len(x)):
     z1=perceptron(x[j],[v[1][0],v[2][0]],v[0][0])
     z2=perceptron(x[j],[v[1][1],v[2][1]],v[0][1])

     y=perceptron([z1,z2],[w[1][0],w[2][0]],w[0][0])

     erout1=(t[j]-y)*y*(1-y)
     e+=(t[j]-y)*(t[j]-y)

     delw1= learnRate * erout1 * z1
     delw2= learnRate * erout1 * z2
     delw0= learnRate * erout1

     errin1=erout1 * w[1][0]
     errin2=erout1 * w[2][0]

     errz1= errin1 * z1 * (1-z1)
     errz2= errin2 * z2 * (1-z2)

     delv11= learnRate * errz1 * x[j][0]
     delv21= learnRate * errz1 * x[j][1]
     delv01= learnRate * errz1
     delv12= learnRate * errz2 * x[j][0]
     delv22= learnRate * errz2 * x[j][1]
     delv02= learnRate * errz2

     v[1][0]=v[1][0] + delv11
     v[1][1]=v[1][1] + delv12
     v[2][0]=v[2][0] + delv21
     v[2][1]=v[2][1] + delv22
     v[0][0]=v[0][0] + delv01
     v[0][1]=v[0][1] + delv02

     w[1][0]=w[1][0] + delw1
     w[2][0]=w[2][0] + delw2
     w[0][0]=w[0][0] + delw0

   if e<=0.002:
      break
   count+=1



# print("number of epochs taken:",count)
# plt.plot(errEpoch)
# plt.show()
```

A4. Learn using a MLP network from Sci-Kit manual available at https://scikitlearn.org/stable/modules/neural_networks_supervised.html. Repeat the AND Gate and XOR Gate
exercises using MLPClassifier() function.

```python
from sklearn.neural_network import MLPClassifier

X = [[0,0],[0,1],[1,1]]

y = [0,0,1]

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)

clf.fit(X, y)
```
