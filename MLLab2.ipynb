A4. Create a “bag-of-words” from the “token population” by removing the stop-words.


```python
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
i=0
bow=[]

#Add elements
while i<len(pop):
  for word in pop[i]: 
    if word not in stops:
      bow.append(word)
  i+=1
print(bow)

#Remove commas,full stops...
for x in bow:
  if not x.isalpha():
    bow.remove(x)

#unique the list
t=set(bow)
bow=list(t)

print("Final Processed Bag of words:")
print(bow)

print('Length:',len(bow))
```

A5. For each document / instance, create 2 feature vectors as follows
o First vector attributes indicate the presence / absence of tokens from bag-of-words
in that document
o Second vector attributes indicate the count of presence of each word from bag-ofwords in the document
o Do this for all documents. After this is done, you should have 37 vectors each of first 
and second kind.

```python
FeatureVect1=[]  #Presence Feature Vector Initialisation
FeatureVect2=[]  #Count Feature Vector Initialisation

it1=0
while it1<len(pop):
   j1=0
   a=[]
   while j1<len(bow):
     booln=bow[j1] in DataFile["Report"][it1]
     if(booln):
       a.append(1)
       j1+=1
     else:
       a.append(0)
       j1+=1
   FeatureVect1.append(a)
   it1+=1

print(FeatureVect1)
print('The count of vectors in Presence Feature Vector:',len(FeatureVect1))



it2=0
while it2<len(pop):
  j2=0
  b=[]
  while j2<len(bow):
    b.append(DataFile["Report"][it2].count(bow[j2]))
    j2+=1
  FeatureVect2.append(b)
  it2+=1

print(FeatureVect2)
print('The count of vectors in Count Feature Vector:',len(FeatureVect2))
```

A6. Calculate the Jaccard Coefficient between the document vectors. Use first vector for each 
document for this. Jaccard coefficient between two binary vectors is defined as:
JC = (f11) / (f01+ f10+ f11)
f11= number of attributes where x was 1 and y was 1; x & y are vectors

```python
JC=[]
for i in range(0,len(pop)):
   temp=[]
   for j in range(0,len(pop)):
     f11=f01=f10=0
     for k in range(0,len(bow)):
       if FeatureVect1[i][k]==1 and FeatureVect1[j][k]==1:
         f11+=1
       elif FeatureVect1[i][k]==0 and FeatureVect1[j][k]==1:
         f01+=1
       elif FeatureVect1[i][k]==1 and FeatureVect1[j][k]==0:
         f10+=1
     temp.append(f11/(f11+f01+f10)) 
   JC.append(temp)
print('Jaccard Coefficient:')
print(JC)
```

A7. Calculate the Cosine similarity between the documents by using the second feature vector for 
each document

```python
import math
CS=[]
for i in range(0,len(pop)):
  temp=[]
  for j in range(0,len(pop)):
    sum=0
    dsq1=dsq2=0
    for k in range(0,len(bow)):
      sum=sum+(FeatureVect2[i][k]*FeatureVect2[j][k])
      dsq1=dsq1+(FeatureVect2[i][k]*FeatureVect2[i][k])
      dsq2=dsq2+(FeatureVect2[j][k]*FeatureVect2[j][k]) 
    temp.append(sum/(math.sqrt(dsq1)*math.sqrt(dsq2))) 
  CS.append(temp)
print('Cosine similarity:')
print(CS)
```
